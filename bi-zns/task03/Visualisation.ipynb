# Fuzzy inference - vizualizace

## Funkce příslušnosti

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
# univerza předpokladů
money = ctrl.Antecedent(np.arange(0, 200, 1), 'Money')
nmb_guards = ctrl.Antecedent(np.arange(0, 28, 1), 'Guards')
place_weak_guards = ctrl.Consequent(np.arange(0, 28, 1), 'Place weak guards')
place_strong_guards = ctrl.Consequent(np.arange(0, 28, 1), 'Place strong guards')
```

```python
# definice funkcí příslušnosti pro předpoklady a závěry
money['poor'] = fuzz.trapmf(money.universe,[0,0,0,70])
money['middle_class'] = fuzz.trapmf(money.universe,[35,80,140,180])
money['rich'] = fuzz.trapmf(money.universe,[150,200, 200, 200])

nmb_guards['few'] = fuzz.trapmf(nmb_guards.universe,[0,0,0,10])
nmb_guards['average'] = fuzz.trapmf(nmb_guards.universe,[8,12,22,26])
nmb_guards['many'] = fuzz.trapmf(nmb_guards.universe,[22,28,28,28])

place_weak_guards['few'] = fuzz.trapmf(place_weak_guards.universe,[0,0,0,10])
place_weak_guards['average'] = fuzz.trapmf(place_weak_guards.universe,[8,12,22,26])
place_weak_guards['many'] = fuzz.trapmf(place_weak_guards.universe,[22,28,28,28])

place_strong_guards['few'] = fuzz.trapmf(place_strong_guards.universe,[0,0,0,10])
place_strong_guards['average'] = fuzz.trapmf(place_strong_guards.universe,[8,12,22,26])
place_strong_guards['many'] = fuzz.trapmf(place_strong_guards.universe,[22,28,28,28])
```

```python
# vizualizace množin 
money.view()
nmb_guards.view()
place_weak_guards.view()
place_strong_guards.view()
```
